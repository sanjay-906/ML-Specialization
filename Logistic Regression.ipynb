```python
from sklearn.linear_model import LogisticRegression
```

```python
import numpy as np

X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1])
```

```python
model=LogisticRegression()
model.fit(X,y)
```

```python
y_pred=model.predict(X)
y_pred
```

```python
model.score(X,y)
```
